```
#use word2vec to build my index of words
model = gensim.models.Word2Vec(
        sentences,
        size=123,
        window=10,
        min_count=2,
        workers=10)
model.train(sentences, total_examples=len(articles), epochs=10)
```

I decided to use convolution layer since it yields the highest accuracy tested above. Compared to LSTM. I tried GRU but it took forever to run, and accuracy was not that high.

```
#my deep network model to be used in 5-fold
def model14():
  model14 = Sequential()
  model14.add(Embedding(10000, 128, input_length=maxlen))
  model14.add(MaxPooling1D(2))
  model14.add(Flatten())
  model14.add(Dense(7, activation='softmax'))
  model14.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['acc'])
  return model14

def model17():
  model17 = Sequential()
  model17.add(Embedding(10000, 128, input_length=maxlen))
  model17.add(MaxPooling1D(2, padding='same'))
  model17.add(GlobalMaxPool1D())
  model17.add(Dense(7, activation='softmax'))
  model17.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
  return model17

def model16():
  model16 = Sequential()
  model16.add(Embedding(10000, 128, input_length=maxlen))
  model16.add(MaxPooling1D(2, padding='same'))
  model16.add(Flatten())
  model16.add(Dense(7, activation='softmax'))
  model16.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['acc'])
  return model16

def model18():
  model18 = Sequential()
  model18.add(Embedding(10000, 256, input_length=maxlen))
  model18.add(MaxPooling1D(2, padding='same'))
  model18.add(GlobalMaxPool1D())
  model18.add(Dense(32, activation='relu'))
  model18.add(Dense(7, activation='softmax'))
  model18.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['acc'])
  return model18
```

```
#this is my best 5-Fold with a 91% accuracy rate! Model17 refer above (40mins to run)
import numpy as np

k = 5
num_val_samples = len(x_train) // k
num_epochs = 30
all_scores = []
for i in range(k):
    print('processing fold #', i)
    # Prepare the validation data: data from partition # k
    val_data = x_train[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = y_train[i * num_val_samples: (i + 1) * num_val_samples]

    # Prepare the training data: data from all other partitions
    partial_train_data = np.concatenate(
        [x_train[:i * num_val_samples],
         x_train[(i + 1) * num_val_samples:]],
        axis=0)
    partial_train_targets = np.concatenate(
        [y_train[:i * num_val_samples],
         y_train[(i + 1) * num_val_samples:]],
        axis=0)

    # Build the Keras model (already compiled)
    model = model17()
    # Train the model (in silent mode, verbose=0)
    model.fit(partial_train_data, partial_train_targets,
              epochs=num_epochs, batch_size=128, verbose=0)
    # Evaluate the model on the validation data
    val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=0)
    all_scores.append(val_mae)
```
